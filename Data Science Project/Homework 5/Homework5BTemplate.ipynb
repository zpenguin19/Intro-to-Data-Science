## 1.1 Get to know data

> #### Data description:
- `Country` - the name of the country that the data is taken from
- `Year` - the year the data of a country is taken from
- `Data.Health.Birth Rate` - the birth rate for the particular country and year
- `Data.Health.Death Rate` - the death rate for the particular country and year
- `Data.Health.Fertility Rate` - the fertility rate for the particular country and year
- `Data.Health.Life Expectancy at Birth, Female` - the life expectancy of a female born in the particular country and year
- `Data.Health.Life Expectancy at Birth, Male` - the life expectancy of a male born in the particular country and year
- `Data.Health.Life Expectancy at Birth, Total`	- the life expectancy for someone born in the particular country and year
- `Data.Health.Population Growth` - the population growth of the country in that year
- `Data.Health.Total Population` - the total population of the particular country in the specified year
- `Data.Infrastructure.Mobile Cellular Subscriptions` - the total number of cell phone subscriptions in the particular country and year
- `Data.Infrastructure.Mobile Cellular Subscriptions per 100 People` - the number of cell phone subscriptions per 100 people in the particular country and year
- `Data.Infrastructure.Telephone Lines` - the total number of telephone lines in the particular country and year
- `Data.Infrastructure.Telephone Lines per 100 People` - the number of telephone lines per 100 people in the particular country and year
- `Data.Rural Development.Agricultural Land` - total area of agricultural land in the particular country and year
- `Data.Rural Development.Agricultural Land Percent` - percent of the country's land that is used for agriculture in the given year
- `Data.Rural Development.Arable Land` - total area of arable land in the particular country and year
- `Data.Rural Development.Arable Land Percent` - percent of the country's land that is arable in the given year
- `Data.Rural Development.Land Area` - total area of rural land in the particular country and year
- `Data.Rural Development.Rural Population` - number of people living in rural areas in the particular country and year
- `Data.Rural Development.Rural Population Growth` - the population growth of rural areas in the particular country in that year
- `Data.Rural Development.Surface Area` - surface area of rural areas in the particular country and year
- `Data.Urban Development.Population Density` - density of people living in urban areas in the particular country and year
- `Data.Urban Development.Urban Population Percent` - percent of the population that lives in urban areas in the given country and year
- `Data.Urban Development.Urban Population Percent Growth` - the population growth of urban areas in the particular country in that year

## 1.2 Identify which numerical columns and categorical columns

```python
# create a list of the columns that are categorical and those that are integer and those that have string values

import numpy as np
numerical = []
categorical = []
string = []

for column in development_df.columns:
  if development_df[column].nunique() < 10:
    categorical.append(column)
  elif development_df[column].dtype == np.float64 or development_df[column].dtype == np.int64:
   numerical.append(column)
  else:
    string.append(column)

print("Numerical columns: ", numerical)
print("Categorical columns: ", categorical)
print("String columns: ", string)

numerical = []
categorical = []
string = []

for column in emissions_df.columns:
  if emissions_df[column].nunique() < 10:
    categorical.append(column)
  elif emissions_df[column].dtype == np.float64 or emissions_df[column].dtype == np.int64:
   numerical.append(column)
  else:
    string.append(column)

print("Numerical columns: ", numerical)
print("Categorical columns: ", categorical)
print("String columns: ", string)
```

##### 1.7.2.1.2 Concatenate DataFrames
We want to be able to work with all of the data we have imported at once, so we need to pull all two DataFrames into one DataFrame. They all have the same column format now, so we can concatenate them based on column name and order (similar to adding them together, one on top of another) using the pandas method concat().

We also need to consider the current index labels for each dataset. We will create a new zero-based integer index label for the concatenated dataset by passing the keyword argument ignore_index=True into the concat() method.

Ensure you are concatenating the cleaned dataframes.

```python
# Concatenate all the datasets into one
development_emissions_df = pd.concat([dev_clean, em_clean], ignore_index=True)


# Print the shape (number of rows and columns) of the full DataFrame
development_emissions_df.shape
```
